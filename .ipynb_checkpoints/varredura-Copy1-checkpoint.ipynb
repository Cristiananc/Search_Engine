```python
import os
import re
import time 
import unidecode as ud

n = os.listdir('Desktop/limpeza2')

t0 = time.time()
def processa(arquivo,t):
    f = open(arquivo,'r',encoding='utf8')
    x = f.read()
    h = x.split("id=")
    x = x.replace('_','')
    h = (ud.unidecode(x)).split("id=")
    v = open('Desktop/3periodo/texto_{}.txt'.format(t),'a',encoding='utf8')
    for i in range(len(h)):
        k = " ".join(re.findall(r'\w+',h[i]))
        h[i] = k
    v.write((" \n  \n id=".join(h).lower()))      
    v.close()
    f.close()

for i in range(len(n) - 1):
    if i == 0:
        y = 0
    else:
        x = re.findall(r'\d+',n[i])
        y = int(x[0])
    processa('Desktop/limpeza2/'+n[i],y)
        
tf = time.time()
print(tf-t0)
```

```python
## Função para classificar os textos originais

import os
import re
import time 
import unidecode as ud
#MARCAO/araw.en
n = os.listdir('../../../Desktop/MARCAO/araw.en')

lista = []
def titulos(arquivo):
    global lista
    f = open(arquivo,'r',encoding = 'utf8')
    x = f.read()
    h = x.split("\n")
    for i in range(len(h)):
        y = h[i].split(" ")
        if len(y) >=4:
            if len(y[1]) >=4 and len(y[2]) >= 7:
                if y[1][0:4] == "id=\"" and y[2][0:7] == "title=\"":
                    for r in range(len(y)):
                        if len(y[r]) >= 12:
                            if y[r][0:12] == 'nonfiltered=':
                                k = r
                                st = y[2][7:]+" "+" ".join(y[3:k])
                                st = st[0:-1]
                                st2 = (ud.unidecode(" ".join(re.findall(r'\w+',st)))).lower()
                                lista.append([st2])
    f.close()
    
def escreve(arquivo,c2):
    global lista
    
    f = open(arquivo,'r',encoding = 'utf8')
    x = f.read()
    h = x.split("<doc ")
    for i in range(len(h)):
        y = h[i].split(" ")
        if len(y) >=3:
            if len(y[0]) >=4 and len(y[1]) >= 7:
                if y[0][0:4] == "id=\"" and y[1][0:7] == "title=\"":
                    for r1 in range(len(y)):
                        if len(y[r1]) >= 12:
                            if y[r1][0:12] == 'nonfiltered=':
                                r = lista[c2][2]
                                y[0] = 'id={}'.format(r)
                                v = open('../../../Desktop/t6/original_{}.txt'.format(int(r/10000)), 'a', encoding = 'utf8')
                                c2 = c2+1 
                                h[i] = " ".join(y)
                                v.write("<doc "+h[i])    
                                v.close()
                        
    f.close()
    return c2

def classifica():
    global lista
    t0 = time.time()
    c2 = 0

    for i in range(len(n)):
        titulos('../../../Desktop/MARCAO/araw.en/'+n[i])
        print(i)
    for i in range(len(lista)):
        lista[i].append(i)
              
    lista.sort()
    print(time.time()-t0)
    for i in range(len(lista)):
        lista[i].append(i)
    
    lista.sort(key = lambda x: x[1])
    for i in range(len(n)):
        c2 = escreve('../../../Desktop/MARCAO/araw.en/'+n[i] ,c2)
        print(i)
    print(time.time()-t0)
    return lista



##função para classificar arquivo a arquivo do código acima

n = os.listdir('../../../Documents/t6')

lista = []
def titulosf(arquivo,t):
    global lista

    f = open(arquivo,'r',encoding = 'utf8')
    x = f.read()
    h = x.split("</doc>\n")
    for i in range(len(h)):
        y = h[i].split(" ")
        if len(y) >=4:
            if len(y[1]) >=4 and len(y[2]) >= 7:
                if y[1][0:3] == "id=" and y[2][0:7] == "title=\"":
                    D = y[1][3:]
                    if D.isdigit:
                        lista.append([int(D),h[i]])

    f.close()
    
def escrevef(t):
    global lista

    lista.sort()
    g = open('../../../Desktop/t6/ord_origin_f{}.txt'.format(t), 'a', encoding = 'utf8')
    for i in range(len(lista)):
        g.write("</doc>\n"+lista[i][1])
        
                        
    g.close()
    
    lista = []

def ordenaf():
    global lista
    
    t0 = time.time()
    c2 = 0

    for i in range(len(n)):
        titulosf('../../../Documents/t6/'+n[i],i)
        escrevef(i)
        print(i)

    print(time.time()-t0)
    return lista
```
