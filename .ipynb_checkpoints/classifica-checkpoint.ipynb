```python
## função para ordenar os textos limpos, similar à anterior

import os
import re
import time 
import unidecode as ud
n = os.listdir('Desktop/3periodo')

t0 = time.time()
c = 0

def le(arquivo,lista):
    global c
    f = open(arquivo,'r',encoding='utf8')
    x = f.read()
    h = x.split("\n")
    for i in range(len(h)):
        y = h[i].split(" ")
        if len(y) >=2:
            if len(y[1]) >=3:
                if y[1][0:3] == 'id=' and y[2] == 'title' and 'nonfiltered'in y:
                    j = y.index('nonfiltered')
                    title = " ".join(y[3:j])
                    lista.append([title,c])
                    c = c+1
    f.close()
    return lista
            
def escreve(arquivo,t,lista,c2):
    
    f = open(arquivo,'r',encoding='utf8')
    x = f.read()
    h = x.split("\n")
    v = open('Desktop/tt4/Texto_{}.txt'.format(t),'a',encoding='utf8')
    for i in range(len(h)):
        y = h[i].split(" ")
        if len(y) >=2:
            if len(y[1]) >=3:
                if y[1][0:3] == 'id=' and y[2] == 'title' and 'nonfiltered'in y:
                    k = y.index('dbindex')
                    j = y.index('nonfiltered')
                    title = " ".join(y[3:j])
                    y = y[0:j]+y[k+2:]
                    y = y[0:j]+['\n']+y[j:]
                    if 'doc' in y:
                        y.remove('doc')
                    if 'doc' in y:
                        y.remove('doc')
                    if 'endofarticle' in y:
                        y.remove('endofarticle')
                    y[1] = 'id={}'.format(lista[c2][2])
                    c2 = c2+1 
                    h[i] = " ".join(y)
    v.write("\n".join(h))      
    v.close()
    f.close()
    return c2

def classifica():
    c2 = 0
    lista = []
    for i in range(len(n)):
        lista = le('Desktop/3periodo/'+n[i],lista)
    
    lista.sort()
    for i in range(len(lista)):
        lista[i].append(i)
    lista.sort(key = lambda x:x[1])
    tf = time.time()-t0
    print(tf)
    for i in range(len(n)):
        x = escreve('Desktop/3periodo/'+n[i],i,lista,c2)
        c2 = x
    print(time.time()-t0)
    
##
```
